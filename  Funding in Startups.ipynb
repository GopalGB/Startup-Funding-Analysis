#### Top 10 Regions by Total Funding for Each Round'

```python
# List of all funding round columns you provided
all_funding_rounds = [
    'seed', 'venture', 'equity_crowdfunding', 'undisclosed', 'convertible_note', 
    'debt_financing', 'angel', 'grant', 'private_equity', 'post_ipo_equity', 
    'post_ipo_debt', 'secondary_market', 'product_crowdfunding', 'round_A', 
    'round_B', 'round_C', 'round_D', 'round_E', 'round_F', 'round_G', 'round_H'
]

# Filter out funding rounds that are actually present in the DataFrame
available_funding_rounds = [round for round in all_funding_rounds if round in startup_data_cleaned.columns]

# Summing the total funding across all available funding rounds
startup_data_cleaned['total_funding'] = startup_data_cleaned[available_funding_rounds].sum(axis=1)

# Sorting the DataFrame by total funding and selecting the top 10 regions
top_10_funding_per_region = startup_data_cleaned.sort_values(by='total_funding', ascending=False).head(10)

# Plotting the line chart for the top 10 regions for each available funding round
# Plotting the line chart for the top 10 regions for each available funding round
plt.figure(figsize=(10, 6))

# Loop through the available funding rounds and plot each
for round_type in available_funding_rounds:
    plt.plot(top_10_funding_per_region['region'], 
             top_10_funding_per_region[round_type], 
             marker='o', 
             label=round_type)

# Adding title and labels
plt.title('Top 10 Regions by Total Funding for Each Round')
plt.xlabel('Region')
plt.ylabel('Funding Amount (USD)')

# Moving the legend outside of the plot
plt.legend(title='Funding Rounds', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjusting x-axis labels and layout
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

```

### **1. Feature Engineering**:
We will create two new features:
- **Funding per Round**: Total funding divided by the number of funding rounds.
- **Startup Age at First Funding**: The difference between the startup's founding year and its first funding year.



```python
# Feature 1: Funding per round (total funding / number of funding rounds)
startup_data_cleaned['funding_per_round'] = startup_data_cleaned['funding_total_usd'] / startup_data_cleaned['funding_rounds']

# Feature 2: Startup age at first funding (first funding year - founded year)
startup_data_cleaned['founded_year'] = pd.to_numeric(startup_data_cleaned['founded_year'], errors='coerce')

# Convert 'first_funding_at' to datetime, invalid parsing will be set as NaT
startup_data_cleaned['first_funding_year'] = pd.to_datetime(startup_data_cleaned['first_funding_at'], errors='coerce').dt.year

# Calculate startup age at the time of first funding
startup_data_cleaned['startup_age_at_funding'] = startup_data_cleaned['first_funding_year'] - startup_data_cleaned['founded_year']

# Display the new features
startup_data_cleaned[['funding_per_round', 'startup_age_at_funding']].head()
```

### Feature Engineering Complete

Two new features have been successfully added to the dataset:

1. **Funding per Round**: This feature represents the average funding per round for each startup. This helps in understanding how much a startup typically raises in each funding round.

2. **Startup Age at First Funding**: This shows the number of years between the startup's founding year and its first funding year. A younger startup (in terms of time from founding to first funding) could indicate stronger growth potential or quicker market traction.

#### Insight:

The funding per round and startup age at first funding features help capture key elements of startup growth and investor interest. Startups with a shorter time to first funding might signal better market readiness or stronger product-market fit, which could be appealing to investors.

### **2. Predictive Modeling**:
We will build a **classification model** to predict whether a startup will receive additional funding based on key features. We will:
- Use a **decision tree** or **random forest** for classification.
- Features used: **region**, **sector (category_list)**, **number of funding rounds**, **funding per round**, and **startup age**.


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Prepare the dataset for predictive modeling
X = startup_data_cleaned[['funding_per_round', 'startup_age_at_funding', 'funding_rounds']]
X = X.fillna(0)  # Fill missing values with 0
y = (startup_data_cleaned['funding_rounds'] > 1).astype(int)  # Success defined as having >1 funding round

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a random forest classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_result = classification_report(y_test, y_pred)

accuracy, classification_report_result
```

### Predictive Modeling Results:

- **Accuracy**: The model achieved an accuracy score of 1.0 (100%) across all folds during cross-validation, indicating perfect predictions. This suggests that the model may have learned the patterns in the training data too well, a sign of overfitting.
  
- **Precision, Recall, and F1-Score**:These metrics were also 1.0 across all classes, further reinforcing that the model is likely overfitted. This performance could be due to the lack of variability in the dataset or because the features have high correlations, leading the model to make easy predictions without being truly challenged.

#### Recommendation:

The model’s performance looks perfect on the training data, but to ensure it generalizes well to unseen data, consider using techniques like feature selection, regularization, or gathering more diverse data. Additionally, reducing multicollinearity (as checked by VIF scores) may improve the model’s ability to generalize.

```python
from sklearn.model_selection import cross_val_score

# Use cross-validation to evaluate the model performance (5-fold cross-validation)
cross_val_scores = cross_val_score(model, X, y, cv=5)

# Display the cross-validation scores and the average score
print("Cross-validation scores: ", cross_val_scores)
print("Average cross-validation score: ", cross_val_scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# Set up the hyperparameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the random forest model
rf_model = RandomForestClassifier(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the model to the training data
grid_search.fit(X_train, y_train)

# Best parameters from the grid search
best_params = grid_search.best_params_

# Evaluate the best model
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)
best_accuracy = accuracy_score(y_test, y_pred_best)
classification_report_best = classification_report(y_test, y_pred_best)

best_params, best_accuracy, classification_report_best
```



### Insights from Hyperparameter Tuning:

The best parameters for the Random Forest model after hyperparameter tuning are:

- **max_depth**: None (no restriction on the depth of the trees).
- **min_samples_leaf**: 1 (each leaf node must have at least one sample).
- **min_samples_split**: 2 (a node is split if it has two or more samples).
- **n_estimators**: 50 (the number of trees in the forest).

Even after tuning the hyperparameters, the model still achieves a perfect accuracy score of **1.0** on both the training and testing sets, which strongly indicates **overfitting**. This may be a result of:

- A **small or homogeneous dataset** that does not introduce enough variability for the model to generalize well.
- **Highly correlated features**, making it easier for the model to make perfect predictions without being sufficiently challenged.

The classification report also reflects perfect precision, recall, and F1-scores of **1.0** across all classes. While this might appear ideal, it further highlights the likelihood of overfitting, as the model is able to predict all outcomes perfectly for the current dataset but may not perform as well on new, unseen data.

To mitigate this, additional strategies such as **feature selection**, **regularization**, or obtaining a more diverse dataset should be considered to improve the model’s generalizability.

### **3. Statistical Testing**:
We will perform **ANOVA** to determine if there are significant differences in **average funding amounts** between different sectors and regions. We’ll start with sectors.


```python
from scipy.stats import f_oneway

# Filter out empty groups for sector funding
sector_groups = [group['funding_total_usd'].dropna() for name, group in startup_data_cleaned.groupby('category_list') if not group['funding_total_usd'].dropna().empty]
# Perform ANOVA for sectors
anova_sectors = f_oneway(*sector_groups)

# Filter out empty groups for region funding
region_groups = [group['funding_total_usd'].dropna() for name, group in startup_data_cleaned.groupby('region') if not group['funding_total_usd'].dropna().empty]
# Perform ANOVA for regions
anova_regions = f_oneway(*region_groups)

# Output the results
anova_sectors, anova_regions
```



### Interpretation of ANOVA Results:

The ANOVA results provide the following insights:

- **For Sectors**: The F-statistic is **0.95** and the p-value is **0.54**.
- **For Regions**: The F-statistic is **0.48** and the p-value is **0.93**.

### What Do These Results Mean?

- The **p-values** for both sectors and regions are **greater than 0.05**, which means there is **no statistically significant difference** in the average funding amounts across different sectors or regions.

  - **Sectors**: The ANOVA test for sectors resulted in an F-statistic of 0.95 and a p-value of 0.54, indicating **no significant difference** in funding across sectors. This suggests that investors are not significantly favoring one sector over another in terms of the average funding amount.

  - **Regions**: Similarly, the ANOVA test for regions resulted in an F-statistic of 0.48 and a p-value of 0.93. This implies that **funding distribution is fairly even** across regions, with no significant difference in the average funding amounts between them.

### Insight:

While certain regions and sectors may appear to attract more funding (e.g., **IoT** and **SF Bay Area**), the statistical analysis reveals that these differences are **not significant at the 0.05 level**, suggesting that funding is more evenly distributed than initially perceived.
